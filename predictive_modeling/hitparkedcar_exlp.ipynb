```python
acc = pd.read_csv('../data/cleaned/coll_coord.csv', low_memory=False)
```

Let's try to predict 'HITPARKEDCAR' by the location.

```python
from sklearn.model_selection import train_test_split
```

```python
X = acc[['longitude','latitude']]
Xf = acc[['HITPARKEDCAR','longitude','latitude','year']]
y_f = np.array(acc[['HITPARKEDCAR']]).reshape(-1,)
```

```python
A = acc[['COLDETKEY','INCDATE','HITPARKEDCAR']]
```

```python
y = np.array([y_f=='Y']).reshape(-1,)
```

```python
years=Xf[y]['year'].unique()
years.sort()
years
```

```python
[sum(Xf[y]['year']==year) for year in years]
```

==> only from 2015 on the "HITPARKEDCAR" column was maintained properly; still the fluctuations are suspicious (data quality probably not good)

- get random sample of this size from the negative part of the data, and merge

```python
Xf_s = Xf[~y].sample(len(X[y]))
```

```python
Xf_b = pd.concat([Xf[y],Xf_s])
Xf_b
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn import tree

dt = tree.DecisionTreeClassifier(criterion='entropy')
dt.fit(X_train, y_train)
```

```python
probabilities = dt.predict_proba(X_test)[:, 1]
probabilities
```

```python
probabilities.max()
```

```python
probabilities.min()
```

```python
sum(dt.predict(X_test)==y_test)/len(y_test)
```

- Decision trees give us a really good result on this.

```python
dt.score(X_test,y_test)
```

```python
dt.score(X,y)
```

```python
dt.tree_.node_count
```

```python
dt.tree_.max_depth
```

Just out of curiosity: how is the predictive power affected if I limit the tree depth? (Guess: dramatically, because the data is highly granular.)

```python
scores=[]
for i in range(1,50):
    dti = tree.DecisionTreeClassifier(criterion='entropy',max_depth=i)
    dti.fit(X_train, y_train)
    scores.append(dti.score(X_test,y_test))
```

```python
plt.plot(range(1,50),scores)
```

```python
dt2 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=1)
dt2.fit(X_train, y_train)
```

```python
dt2.score(X_test,y_test)
```

```python
dt2.tree_.node_count
```
