```python
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

x = range(2004,2019)
y = range(2004,2019)
z = np.zeros(15)

dx = np.ones(15)
dy = np.ones(15)
dz = range(1,16)

ax1.bar3d(x,y,z,dx,dy,dz)
```
